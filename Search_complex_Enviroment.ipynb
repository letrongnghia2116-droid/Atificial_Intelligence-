```python
import copy

GOAL = (
    (1, 2, 3),
    (4, 5, 6),
    (7, 8, 0)
)


# belief_state là một 'frozenset' chứa các tuple trạng thái khả thi
class BeliefNode:
    def __init__(self, belief_state, parent=None, action=None):
        self.belief_state = frozenset(belief_state)
        self.parent = parent
        self.action = action


def find_blank(state):
    for i in range(3):
        for j in range(3):
            if state[i][j] == 0:
                return i, j
    return None


def result(state, action):
    x, y = find_blank(state)

    if action == 'up' and x > 0:
        nx, ny = x - 1, y
    elif action == 'down' and x < 2:
        nx, ny = x + 1, y
    elif action == 'left' and y > 0:
        nx, ny = x, y - 1
    elif action == 'right' and y < 2:
        nx, ny = x, y + 1
    else:
        return state

    new_state = [list(row) for row in state]
    new_state[x][y], new_state[nx][ny] = new_state[nx][ny], new_state[x][y]
    return tuple(tuple(row) for row in new_state)


def transition_belief(belief_state, action):
    new_belief = set()
    for state in belief_state:
        next_state = result(state, action)
        new_belief.add(next_state)
    return frozenset(new_belief)


def generate_all_possible_states():
    state1 = ((1, 2, 3), (4, 0, 6), (7, 5, 8))
    state2 = ((1, 2, 3), (4, 5, 6), (0, 7, 8))
    state3 = ((1, 2, 3), (4, 5, 0), (7, 8, 6))
    return frozenset([state1, state2, state3])

#Tìm kiếm trên không gian trạng thái niềm tin để gom tất cả các nguồn về một Đích duy nhất
def Belief_Graph_Search(initial_belief, goal_state):
    frontier = []
    root = BeliefNode(initial_belief, None, None)
    frontier.append(root)

    reached = set()
    reached.add(root.belief_state)

    actions_list = ['up', 'down', 'left', 'right']

    while frontier:
        node = frontier.pop(0)
        if len(node.belief_state) == 1 and goal_state in node.belief_state:
            return node

        for action in actions_list:
            next_belief = transition_belief(node.belief_state, action)

            if next_belief not in reached:
                reached.add(next_belief)
                child = BeliefNode(next_belief, node, action)
                frontier.append(child)

    return None


def solution(node):
    path = []
    while node.parent is not None:
        path.append(node.action)
        node = node.parent
    path.reverse()
    return path


def print_state(state):
    for row in state:
        print(row)
    print()


def main():

    initial_belief = generate_all_possible_states()

    print("=== BÀI TOÁN MÔI TRƯỜNG PHỨC TẠP KHÔNG BIẾT NGUỒN ===")
    print(f"Số lượng trạng thái nguồn có thể xảy ra: {len(initial_belief)}\n")
    for idx, st in enumerate(initial_belief):
        print(f"Khả năng nguồn {idx + 1}:")
        print_state(st)

    print("Trạng thái Đích duy nhất cần đạt:")
    print_state(GOAL)

    print("Đang tính toán chuỗi hành động mù (Conformant Planning)...")
    goal_node = Belief_Graph_Search(initial_belief, GOAL)

    if goal_node is None:
        print("Không tìm thấy chuỗi hành động phù hợp.")
        return

    path = solution(goal_node)
    print("\n Successful!")
    print("Số bước dịch chuyển:", len(path))
    print("Chuỗi hành động đồng bộ:", path)
    print("\n==========================================")


if __name__ == "__main__":
    main()
```
